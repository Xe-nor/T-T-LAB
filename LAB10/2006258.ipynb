```python
df['salary'] = df.groupby('service')['salary'].transform(lambda x: x.fillna(x.mean()))
```

```python
grouped = df.groupby('service')
service_means = grouped['salary'].mean()
df['salary'].fillna(df['service'].map(service_means), inplace=True)
print(df)
```

```python
df['phd'] = df.groupby('service')['phd'].transform(lambda x: x.fillna(x.mean()))
```

```python
grouped = df.groupby('service')
service_means = grouped['phd'].mean()
df['phd'].fillna(df['service'].map(service_means), inplace=True)
print(df)
```

```python
import pandas as pd
from sklearn.linear_model import LinearRegression

data = pd.DataFrame({
    'Duration': [30, 30, 45, 45, 45, 60, 60, 60, 75, 75],
    'Average Pulse': [80, 85, 90, 95, 100, 105, 110, 115, 120, 125],
    'Max Pulse': [120, 120, 130, 130, 140, 140, 145, 145, 150, 150],
    'Caloric Damage': [240, 250, 260, 270, 280, 290, 300, 310, 320, 330],
    'Hours Work': [10, 10, 8, 8, 0, 7, 7, 8, 0, 8],
    'Hours Sleep': [7, 7, 7, 7, 7, 8, 8, 8, 8, 8],
    })

X = data[['Average Pulse']]
y = data['Caloric Damage']

model = LinearRegression()
model.fit(X, y)

new_patient = pd.DataFrame({'Average Pulse': [110]})
predicted_calories = model.predict(new_patient)

print('Predicted Caloric Damage:', predicted_calories[0])
```

```python
import pandas as pd
from sklearn.linear_model import LinearRegression

data = pd.DataFrame({
    'Height (IN CMS)': [158, 158, 160, 160, 163, 163, 165, 165, 168, 170],
    'WEIGHT IN KO)': [58, 62, 60, 59, 61, 60, 61, 65, 66, 64],
    'T-SHIRT SIZE': ['M', 'M', 'M', 'M', 'M', 'M', 'L', 'L', 'L', 'L'],
    })

data['T-SHIRT SIZE'] = data['T-SHIRT SIZE'].replace({'M': 0, 'L': 1})

X = data[['Height (IN CMS)', 'WEIGHT IN KO)']]
y = data['T-SHIRT SIZE']

model = LinearRegression()
model.fit(X, y)

new_customer = pd.DataFrame({'Height (IN CMS)': [170], 'WEIGHT IN KO)': [70]})
predicted_tshirt = model.predict(new_customer)

if predicted_tshirt[0] == 0:
    print('Predicted T-shirt size: M')
else:
    print('Predicted T-shirt size: L')
```
